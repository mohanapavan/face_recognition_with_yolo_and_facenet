```python
!pip install ultralytics
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from pathlib import Path
import csv
import os
```

```python
data_yaml = """
path: /kaggle/input/synthetic-2-real-object-detection-challenge-2/Synthetic to Real Object Detection Challenge 2

train: /kaggle/working/Face-Detection-25/train/images
  
val: /kaggle/working/Face-Detection-25/valid/images
  
nc: 1
names: ['face']
"""

# DosyayÄ± kaydet
with open('data.yaml', 'w') as file:
    file.write(data_yaml)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")

results = model.train(
    data="data.yaml",
    epochs=30,                  # Increased for convergence (early stopping will handle overfitting)
    imgsz=960,                  # High resolution for small objects
    batch=16,         # Early stopping if no improvement
    cos_lr=True,                # Cosine learning rate scheduler
    dropout=0.4,                # Regularization to prevent overfitting
    mosaic=0.4,                 # More aggressive augmentation (50% chance)
    lr0=0.001,
    lrf=0.001,        # Initial LR (higher for SGD)
    optimizer="SGD",          # AdamW > SGD for adaptive learning rates
    momentum=0.937,
    weight_decay=0.0005,
    single_cls=True, 
    plots=True,
    cache=True,
    device="0",                 # Use GPU 0
    workers=8,
    flipud=0.3,
    fliplr=0.3,
    save_period=10
)
```
